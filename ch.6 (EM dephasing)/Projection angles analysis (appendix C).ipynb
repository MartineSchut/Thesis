```python
import numpy as np
import math
import matplotlib.pylab as plt
```

## Charged interferometer

```python
a=3/2
```

```python
c = 0.1
x = y = np.arange(0, math.pi, 0.01)
X, Y = np.meshgrid(x, y)
Z = (np.cos(X) + c * np.cos(Y)) / (1 + c**2)**a

fig, ax = plt.subplots()
im = ax.imshow(abs(Z), interpolation='gaussian',
               origin='lower', extent=[0, math.pi, 0, math.pi],
               vmax=Z.max(), vmin=abs(Z).min())
ax.set_xlabel(r'$\alpha$',size=18)
ax.set_ylabel(r'$\beta$',size=18)
ax.tick_params(axis='both', which='major', labelsize=16)
#cbar = plt.colorbar(im)
#cbar.set_ticks([])
cbar = fig.colorbar(im, ticks=[0.1, 1])
cbar.ax.set_yticklabels([r'low $a_x$', r'high $a_x$'],fontsize=18)
plt.tight_layout()
#plt.savefig('angles-charged-01.png', dpi=800)
plt.show()
```

## Neutral interferometer

```python
c = 0.1
x = y = np.arange(0, math.pi, 0.01)
X, Y = np.meshgrid(x, y)
Gamma = np.sqrt(1-np.cos(Y)**2)
Theta0 = np.sqrt(1-np.cos(X)**2)
Z = ((np.cos(X) + c * np.cos(Y))/ (1 + c**2)**(5/2)) * (Theta0 + c * Gamma)

fig, ax = plt.subplots()
im = ax.imshow(abs(Z), interpolation='gaussian',
               origin='lower', extent=[0, math.pi, 0, math.pi],
               vmax=Z.max(), vmin=abs(Z).min())
ax.set_xlabel(r'$\alpha$',size=16)
ax.set_ylabel(r'$\beta$',size=16)
ax.tick_params(axis='both', which='major', labelsize=14)
# cbar = plt.colorbar(im)
# cbar.set_ticks([])
cbar = fig.colorbar(im, ticks=[0.05, 0.55])
cbar.ax.set_yticklabels([r'low $a_x$', r'high $a_x$'],fontsize=16)
plt.tight_layout()
plt.savefig('angles-neutral-01.png', dpi=800)
plt.show()
```

### Dipole-dipole

```python
c = 0.1
x = y = np.arange(0, math.pi, 0.01)
X, Y = np.meshgrid(x, y)
Gamma = np.sqrt(1-np.cos(Y)**2)
Theta0 = np.sqrt(1-np.cos(X)**2)
Z = ((np.cos(X) + c * np.cos(Y))/ (1 + c**2)**(5/2)) * (Theta0)

fig, ax = plt.subplots()
im = ax.imshow(abs(Z), interpolation='gaussian',
               origin='lower', extent=[0, math.pi, 0, math.pi],
               vmax=Z.max(), vmin=abs(Z).min())
ax.set_xlabel(r'$\alpha$',size=16)
ax.set_ylabel(r'$\beta$',size=16)
ax.tick_params(axis='both', which='major', labelsize=14)
# cbar = plt.colorbar(im)
# cbar.set_ticks([])
cbar = fig.colorbar(im, ticks=[0.05, 0.52])
cbar.ax.set_yticklabels([r'low $a_x$', r'high $a_x$'],fontsize=16)
plt.tight_layout()
plt.savefig('angles-dipole-dipole-01.png', dpi=800)
plt.show()
```
